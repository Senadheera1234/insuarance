# Data Preprocessing: Encoding and Normalization

Preparing the dataset for model training by encoding labels and normalizing numerical features.

```python
# Encode the labels
label_encoder = LabelEncoder()
music_df['label'] = label_encoder.fit_transform(music_df['label'])

# Separate features and target
X = music_df.drop(columns=["label", "filename"])  # Features
y = music_df['label']  # Target (genres)

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Splitting the Dataset

The dataset is divided into training and testing subsets to evaluate the model's performance on unseen data.

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Training and Evaluating Models

## Random Forest

```python
# Initialize and train the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

The confusion matrix for the SVM classifier shows high accuracy for genres like classical and metal, with most predictions correctly classified. ## Hyperparameter Tuning for SVM

```python
# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}
# Initialize GridSearchCV
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)

# Fit the model
grid.fit(X_train, y_train)

# Display the best parameters
print("Best Parameters:", grid.best_params_)
```

```python
# Predict using the best model
y_pred_best_svm = grid.best_estimator_.predict(X_test)

# Evaluate the tuned model
print("Tuned SVM Accuracy:", accuracy_score(y_test, y_pred_best_svm))
print("\nTuned SVM Classification Report:\n", classification_report(y_test, y_pred_best_svm))
```

```python
# Visualize results of grid search
results = grid.cv_results_
scores = results['mean_test_score'].reshape(len(param_grid['C']), len(param_grid['gamma']))

# Plot heatmap of scores
plt.figure(figsize=(8, 6))
sns.heatmap(scores, annot=True, xticklabels=param_grid['gamma'], yticklabels=param_grid['C'], cmap="coolwarm")
plt.title("Grid Search Scores for C and Gamma")
plt.xlabel("Gamma")
plt.ylabel("C")
plt.show()
```

The optimal SVM hyperparameters are C = 10 Gamma = 0.01
These parameters achieve the highest grid search score of 0.89, indicating a strong balance between model complexity and generalization.

## Interpretation of the Scatterplot: Petal Width (cm) vs Target
This scatterplot shows the relationship between the petal width of flowers and their corresponding target class (species) in the Iris dataset.
Petal width (cm) is a critical feature for classifying species in the Iris dataset. It perfectly separates Class 0 (target = 0) from the other two classes (Class 1 and Class 2), demonstrating its strong discriminatory power. While there is minor overlap between Class 1 and Class 2 in the range of 1.5 to 2.0 cm, petal width remains highly effective for distinguishing these classes when combined with other features. Its ability to provide significant separation with minimal complexity makes it an essential feature for improving classification accuracy and should be prioritized in feature selection and model training.

## Import Necessary Libraries

```python
# Import Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
```

## Data Preprocessing

```python
music_df2 = pd.read_csv("features_30_sec.csv")
```

```python
music_df2.describe()
```

```python
music_df2.head()
```

```python
music_df2.isnull().sum()
```

```python
# Drop unnecessary columns
data_cleaned = music_df2.drop(columns=["filename", "length"])

# Encode target labels
label_encoder = LabelEncoder()
data_cleaned["label"] = label_encoder.fit_transform(data_cleaned["label"])

# Separate features and target
X = data_cleaned.drop(columns=["label"])
y = data_cleaned["label"]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Spliting the Dataset 

```python
# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

## Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(8, 6))
sns.countplot(x=y, palette="pastel")
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Frequency")
plt.show()
```

The dataset has a perfectly balanced class distribution, with each class containing 100 samples. This ensures fair training, prevents model bias, and allows reliable evaluation of metrics like accuracy, precision, and recall. It provides a strong foundation for training machine learning models.

## Random Forest Model

```python
# Initialize Random Forest model
rf_model = RandomForestClassifier(random_state=42)
```

## Hyperparameter tuning

```python
# Define hyperparameter space
rf_param_space = {
    "n_estimators": [50, 100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
# Perform hyperparameter tuning with RandomizedSearchCV
random_search_rf = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=rf_param_space,
    n_iter=10,  # Number of combinations to try
    scoring="accuracy",
    cv=3,
    random_state=42,
    n_jobs=-1,
    verbose=1
)
random_search_rf.fit(X_train, y_train)
```

## Evaluate the Best Model

```python
# Best Random Forest model
best_rf_model = random_search_rf.best_estimator_

# Cross-validation scores
cv_scores = cross_val_score(best_rf_model, X_train, y_train, cv=5, scoring="accuracy")
print(f"Cross-Validation Accuracy: {cv_scores.mean()}")

# Predict on test data
y_pred_rf = best_rf_model.predict(X_test)

# Evaluate the model
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf, average="weighted")
rf_auc = roc_auc_score(y_test, best_rf_model.predict_proba(X_test), multi_class="ovr")

rf_classification_report = classification_report(y_test, y_pred_rf, target_names=label_encoder.classes_)
```

```python
print("Best Parameters:", random_search_rf.best_params_)
print("Test Accuracy:", rf_accuracy)
print(f"F1 Score: {rf_f1}")
print(f"ROC-AUC: {rf_auc}")
print("Classification Report:\n", rf_classification_report)
```

## Visualizations

```python
#1 Confusion Matrix Heatmap
conf_matrix = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

The classifier performs well for distinct genres like classical and metal but struggles with overlapping genres like disco and reggae, resulting in misclassifications. Improving feature extraction, dimensionality reduction, or model tuning could enhance performance for these challenging classes.
